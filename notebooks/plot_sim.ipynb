```python
import os

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sim_folder = '/Users/mpages/Projects/cftf/demo'
sim_file = os.path.join(sim_folder, 'test10000.txt.sim')
```

```python
sim_results = pd.read_csv(sim_file, sep = '\t', header = 0, index_col = None)
x = np.array(sim_results)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 5))

ax.imshow(x[:, :60], cmap = 'Greys')

ax.set_ylabel('Tumor fraction (%)')
ax.set_yticks(np.arange(0, 20, 1))
ax.set_yticklabels(np.arange(0, 20, 1))

ax.set_xlabel('Detected alternative alleles')
ax.set_xticks(np.arange(0, 60, 5))
ax.set_xticklabels(np.arange(0, 600, 50))

ax.set_title('10.000 simulations per tumor fraction')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (3, 5))
ax.barh(np.arange(0, 20), x[:, 20][::-1]/100, color = 'black')

ax.set_ylabel('Tumor fraction (%)')
ax.set_yticks(np.arange(0, 20, 1))
ax.set_yticklabels(np.arange(0, 20, 1)[::-1])

ax.set_xlabel('Percentage of simulations')
fig.tight_layout()
plt.show()
```
